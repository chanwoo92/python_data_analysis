## 공공데이터포털 xml 데이터 수집(API)

```python
import requests
from lxml import html
import bs4
```

```python
aptTradeUrl = "http://apis.data.go.kr/1613000/RTMSDataSvcAptTradeDev/getRTMSDataSvcAptTradeDev?serviceKey=AT33s775KYpJOkUBJu0dxkJuUeIfDIOJRzAH084EQS3JN%2BzFjErLHuk%2FGZa9L4gBTSGCzeA69tI9PwLp7B37IQ%3D%3D&LAWD_CD=11500&DEAL_YMD=202407"
```

```python
response = requests.get(aptTradeUrl)
```

```python
xml = response.text
```

```python
xmlObj = bs4.BeautifulSoup(xml, "lxml-xml")
```

```python
targetTag = "items"
itemsTag = xmlObj.find(name=targetTag)
finalTargetTag = "item"
itemTags = itemsTag.findAll(name=finalTargetTag)
```

```python
rowList = []
columnList = []
```

```python
for j in range(0, len(itemTags)):
  eachRow = itemTags[j]
  allColumns = eachRow.findAll()
  for i in range(0, len(allColumns)):
    eachColumn = allColumns[i]
    eachColumnTextValue = eachColumn.text
    columnList.append(eachColumnTextValue)
  rowList.append(columnList)
  columnList = []
```
