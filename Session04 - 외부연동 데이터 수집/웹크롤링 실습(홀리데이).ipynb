```python
import requests, bs4
import pandas as pd

#웹페이지 html 소스 가져오기
resp = requests.get("https://www.timeanddate.com/holidays/south-korea/2023?hol=4194329")
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
bs
```

```python
tabletag = bs.find("table", {"id":"holidays-table"})
tabletag
```

```python
rows = tabletag.find("tbody").findAll("tr")
rows
```

```python
rowList=[]
columnList =[]
```

```python
for i in range(0, len(rows)):
    columns = rows[i].findAll("td")
    columnsLen = len(columns)
    for j in range(0, columnsLen):
        columnList.append(columns[j].text)

    rowList.append(columnList)
    columnList=[]
```

```python
rowList
```

```python
pd.DataFrame(rowList, columns =["Date","Name","Type"])
```
